```python
# eda.ipynb

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
patients = pd.read_csv("../data/patients.csv")
visits = pd.read_csv("../data/hospital_visits.csv")

# Merge for EDA
df = pd.merge(visits, patients, on="patient_nbr", how="left")

# Basic stats
print(df.shape)
print(df["readmitted"].value_counts())

# Plot readmission rate
sns.countplot(data=df, x="readmitted")
plt.title("Readmission Distribution")
plt.show()

# Gender vs readmission
sns.countplot(data=df, x="gender", hue="readmitted")
plt.title("Readmission by Gender")
plt.show()

# Age vs readmission
sns.countplot(data=df, x="age", hue="readmitted")
plt.title("Readmission by Age")
plt.xticks(rotation=45)
plt.show()
```
